```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = 'G:\\ltyworkstuff\\postdoc\\Results\\TRGAN_Michele_NEI_LR\\output\\TRGAN_Michele_NEI64_KW1_SW1_NGO32_LR0.00002-150-11-0.5_AR1.e-1_VG0_OL0_GS2\\train_history.csv'
df_train_history = pd.read_csv(filename)
```

```python
boundaries = 150 + np.arange(11) * 150
```

### *Plot the reconstruction loss $\mathcal{L}_{rec}$.*

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.xlabel('epoch', fontsize=11)
plt.ylabel('$\mathcal{L}_{rec}$', fontsize=14)
plt.title('(b) global random seed = 2')

plt.plot(df_train_history.index+1, df_train_history['mse_loss'], label='training data')
plt.plot(df_train_history.index+1, df_train_history['val_mse_loss'], label='validation data')

plt.grid(axis='y')
#plt.legend(loc='upper right', fontsize='large')

ymin, ymax = plt.gca().get_ylim()

for boundary in boundaries:
    plt.plot([boundary+1, boundary+1], [ymin, ymax], '--k', lw=1)

plt.ylim(ymin, ymax)
```

### *Plot the adversarial loss $\mathcal{L}_{adv}$.*

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.xlabel('epoch', fontsize=11)
plt.ylabel('$\mathcal{L}_{adv}$', fontsize=14)
plt.title('(b) global random seed = 2')

plt.plot(df_train_history.index+1, df_train_history['adversarial_loss'], label='training data')
plt.plot(df_train_history.index+1, df_train_history['val_adversarial_loss'], label='validation data')

# The line $\mathcal{L}_{adv}=-\mathrm{ln}(0.5)$
plt.plot([1,1800], [-np.log(0.5),-np.log(0.5)], '--k', label='$\mathcal{L}_{adv}=-\mathrm{ln}(0.5)$')

plt.grid(axis='y')
#plt.legend(loc='upper right', fontsize='large')

ymin, ymax = plt.gca().get_ylim()

for boundary in boundaries:
    plt.plot([boundary+1, boundary+1], [ymin, ymax], '--k', lw=1)

plt.ylim(ymin, ymax)

#plt.annotate('(a)', (0.15, 0.9), xycoords='axes fraction', fontsize=12)
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.xlabel('epoch', fontsize=11)
plt.ylabel('$\mathcal{L}_{adv}$', fontsize=14)
plt.title('(b) global random seed = 2')

plt.plot(df_train_history.index[900:]+1, df_train_history['adversarial_loss'][900:], label='training data')
plt.plot(df_train_history.index[900:]+1, df_train_history['val_adversarial_loss'][900:], label='validation data')

# The line $\mathcal{L}_{adv}=-\mathrm{ln}(0.5)$
plt.plot([901,1800], [-np.log(0.5),-np.log(0.5)], '--k', label='$\mathcal{L}_{adv}=-\mathrm{ln}(0.5)$')

plt.grid(axis='y')

ymin, ymax = plt.gca().get_ylim()

for boundary in boundaries[5:]:
    plt.plot([boundary+1, boundary+1], [ymin, ymax], '--k', lw=1)

plt.ylim(ymin, ymax)

#plt.annotate('(b)', (0.15, 0.9), xycoords='axes fraction', fontsize=12)
```
